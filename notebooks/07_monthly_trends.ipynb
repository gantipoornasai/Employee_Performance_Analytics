```python
# ── Generate Monthly KPI Trend Data ──

import pandas as pd
import numpy as np
import os

notebook_dir = os.getcwd()
project_root = os.path.dirname(notebook_dir)

np.random.seed(42)

# ── Define 12 months ──
months = [
    '2024-01', '2024-02', '2024-03', '2024-04',
    '2024-05', '2024-06', '2024-07', '2024-08',
    '2024-09', '2024-10', '2024-11', '2024-12'
]

# ── Generate all KPI columns ──
monthly_kpis = pd.DataFrame({
    'Month': months,

    'Avg_Performance_Rating': np.clip(
        3.0 + np.linspace(0, 0.15, 12)
        + np.random.normal(0, 0.05, 12),
        1, 5
    ).round(2),

    'Pct_High_Performers': np.clip(
        24.0 + np.linspace(0, 4.0, 12)
        + np.random.normal(0, 0.8, 12),
        0, 100
    ).round(1),

    'Avg_OKR_Completion': np.clip(
        74.0 + np.array([0,-1,-2,-1,0,1,2,3,2,3,4,5])
        + np.random.normal(0, 1.0, 12),
        0, 100
    ).round(1),

    'Avg_Engagement_Score': np.clip(
        57.0 + np.array([0,0,-1,-2,-1,0,1,1,2,2,3,4])
        + np.random.normal(0, 1.5, 12),
        0, 100
    ).round(1),

    'Promotion_Ready_Count': np.clip(
        42 + np.linspace(0, 10, 12)
        + np.random.normal(0, 1.5, 12),
        0, 500
    ).round(0).astype(int),

    'Flight_Risk_Count': np.clip(
        45 + np.array([0,2,1,-1,-2,-3,-4,-3,-4,-5,-5,-6])
        + np.random.normal(0, 1.0, 12),
        0, 200
    ).round(0).astype(int),

    'Pct_Burnout_High': np.clip(
        22.0 + np.array([0,1,2,1,0,-1,-1,-2,-2,-3,-3,-4])
        + np.random.normal(0, 0.8, 12),
        0, 100
    ).round(1),

    'PIP_Risk_Count': np.clip(
        38 + np.array([0,1,0,-1,-2,-2,-3,-3,-4,-4,-5,-5])
        + np.random.normal(0, 1.0, 12),
        0, 200
    ).round(0).astype(int),

    'Avg_360_Score': np.clip(
        3.1 + np.linspace(0, 0.1, 12)
        + np.random.normal(0, 0.03, 12),
        1, 5
    ).round(2),

    'Avg_Absenteeism_Days': np.clip(
        4.2 + np.array([0,0.1,0.2,0.3,0.2,0.1,
                         0,-0.1,-0.1,-0.2,-0.2,-0.3])
        + np.random.normal(0, 0.1, 12),
        0, 30
    ).round(1),
})

# ── Prior year comparison columns ──
monthly_kpis['PY_Avg_Performance_Rating'] = (
    monthly_kpis['Avg_Performance_Rating']
    - np.random.uniform(0.05, 0.15, 12)
).round(2)

monthly_kpis['PY_Pct_High_Performers'] = (
    monthly_kpis['Pct_High_Performers']
    - np.random.uniform(1, 3, 12)
).round(1)

monthly_kpis['PY_Avg_OKR_Completion'] = (
    monthly_kpis['Avg_OKR_Completion']
    - np.random.uniform(1, 4, 12)
).round(1)

monthly_kpis['PY_Avg_Engagement_Score'] = (
    monthly_kpis['Avg_Engagement_Score']
    - np.random.uniform(2, 5, 12)
).round(1)

monthly_kpis['PY_Flight_Risk_Count'] = (
    monthly_kpis['Flight_Risk_Count']
    + np.random.uniform(3, 8, 12)
).round(0).astype(int)

# ── Target columns ──
monthly_kpis['Target_Avg_Rating']       = 3.5
monthly_kpis['Target_OKR_Completion']   = 75.0
monthly_kpis['Target_Engagement']       = 65.0
monthly_kpis['Target_High_Perf_Pct']    = 30.0
monthly_kpis['Target_Burnout_High_Pct'] = 15.0

# ── Variance columns ──
monthly_kpis['Var_Rating_vs_PY'] = (
    monthly_kpis['Avg_Performance_Rating']
    - monthly_kpis['PY_Avg_Performance_Rating']
).round(3)

monthly_kpis['Var_OKR_vs_PY'] = (
    monthly_kpis['Avg_OKR_Completion']
    - monthly_kpis['PY_Avg_OKR_Completion']
).round(1)

monthly_kpis['Var_Engagement_vs_PY'] = (
    monthly_kpis['Avg_Engagement_Score']
    - monthly_kpis['PY_Avg_Engagement_Score']
).round(1)

monthly_kpis['Var_HP_vs_PY'] = (
    monthly_kpis['Pct_High_Performers']
    - monthly_kpis['PY_Pct_High_Performers']
).round(1)

# ── Save to CSV ──
output_path = os.path.join(
    project_root, 'data', 'processed',
    'monthly_kpi_trends.csv'
)

os.makedirs(os.path.dirname(output_path), exist_ok=True)
monthly_kpis.to_csv(output_path, index=False)

# ── Verify ──
print("-" * 55)
print("monthly_kpi_trends.csv GENERATED")
print("-" * 55)
print(f"\nRows    : {len(monthly_kpis)} months")
print(f"Columns : {len(monthly_kpis.columns)}")
print(f"\nSaved to:")
print(f"{output_path}")
print(f"\nAll columns created:")
for i, col in enumerate(monthly_kpis.columns, 1):
    print(f"{i:>2}. {col}")
print(f"\nPreview:")
print(monthly_kpis[[
    'Month',
    'Avg_Performance_Rating',
    'Pct_High_Performers',
    'Avg_OKR_Completion',
    'Avg_Engagement_Score',
    'Flight_Risk_Count'
]].to_string(index=False))
```
